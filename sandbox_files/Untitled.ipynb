```python
for i in range (0,len(symbols)):
    counter = i
    symbol = symbols[i]       

    
    print("counter is ", counter,"symbol is ",symbol)
    try:
        sleep(0.03)
        start_time = datetime(2014, 1, 1)
        end_time = datetime(2019, 6, 4)
#         df =  get_historical_data(symbol, start, end,output_format='pandas')
        df = pdr.get_data_yahoo(symbol,start_time,end_time)
        file_name = symbol+".csv"
        file_path  = os.path.join(os.getcwd(),"stocks_data",file_name)
#         file_path = os.path.join(os.getcwd(),file_name)
        print(file_path)
        df.to_csv(file_path)
    except:
        print("Error occurred")
        failed_names.append(symbol)
        sleep(3)
```

```python
def missing_files():
    present_files = os.listdir(os.path.join(os.getcwd(),"stocks_data"))
    for i in range(0,len(present_files)):
        file = present_files[i]
        file = file[:-4]
        present_files[i] = file       
    

    set1 = set(symbols)
    set2 = set(present_files)
    
    return list(set1 - set2)
    
```
